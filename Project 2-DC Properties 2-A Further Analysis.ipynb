```python
import pandas as pd
import numpy as np

df = pd.read_csv("datasets/DC_Properties.csv")
df.head(10)
#df.tail(3)
```

```python
#Basic Stats of File
df.describe()
```

```python
#We decided to look at number of Bedrooms, Price, Number of Bathrooms, Year Built to see which home we should buy. 
#We decided to create different Data Frames for each of the four quadrants in DC. 
#We will need to map the data out a little to determine the best house for our needs. 
#Creating Data Frame for DC Homes in NE
df_NE = pd.DataFrame(data=df, columns=['BATHRM','BEDRM', 'PRICE','AYB','QUADRANT'])
df_NE=df_NE.loc[(df_NE['QUADRANT']=='NE')]
df_NE = df_NE.dropna()
df_NE.head()
```

```python
#Creating Data Frame for DC Homes in NW
df_NW = pd.DataFrame(data=df, columns=['BATHRM','BEDRM',"PRICE", 'AYB','QUADRANT'])
df_NW=df_NW.loc[(df_NW['QUADRANT']=='NW')]
df_NW = df_NW.dropna()
df_NW.head()
```

```python
#Creating Data Frame for DC Homes in NW
df_SE = pd.DataFrame(data=df, columns=['BATHRM','BEDRM', "PRICE",'AYB','QUADRANT'])
df_SE=df_SE.loc[(df_SE['QUADRANT']=='SE')]
df_SE = df_SE.dropna()
df_SE.head()
```

```python
#Creating Data Frame for DC Homes in NW
df_SW = pd.DataFrame(data=df, columns=['BATHRM','BEDRM', 'PRICE', 'AYB','QUADRANT'])
df_SW=df_SW.loc[(df_SW['QUADRANT']=='SW')]
df_SW = df_SW.dropna()
#df_SW.rename(columns={index: 'House Id Number'}, inplace =True)
#df_SW.set_index('House Id Number', inplace=True)
df_SW.head()
```

```python
#add more details to scatterplot

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

df_SW=df_SW.drop(df_SW[df_SW['BEDRM']==0].index)



sns.lmplot(x='BEDRM', y='PRICE', data=df_SW, hue='BATHRM')
```

```python
sns.lmplot(x='AYB', y="PRICE", data=df_SW, 
           fit_reg=False #remove regression line
           )   #color by evolution stage
```

```python
df_SW.head()
```

```python

%matplotlib inline

sns.boxplot(data=df_SW)
```

```python
#Define bins as 0-500000, 500001-1000000, 1000001 -1500000, 1500001-2000000, 2000001-2500000
bins = [0,500000, 1000000, 1500000, 2000000, 2500000]

# Create names for the four groups
group_names = ['A', 'B', 'C', 'D', 'E']

#make new column with letter grades
df_SW['SW_Price_bucket'] = pd.cut(df_SW['PRICE'], bins,labels=group_names)
df_SW.head()
```

```python
#SW Price of each Bedroom
df_SW.corr()
```

```python
import statsmodels.formula.api as smf

result = smf.ols('PRICE ~ BATHRM + BEDRM + AYB - 1', data=df_SW).fit()
result.summary()
```
